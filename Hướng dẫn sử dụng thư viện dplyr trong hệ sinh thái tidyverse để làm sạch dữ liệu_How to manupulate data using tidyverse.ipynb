> Take the penguins dataset and then filter for all penguins that live on Torgersen Island

* _Take the penguins dataset_ **translates to** `penguins`

### Creating a subset

It's a little hard to see what's going on in the above table, so I'm going to create a smaller subset of the `penguins` dataset so that we can see what's going on a bit more clearly. You can run the code on the subset of the data, or replace `penguins_subset` with `penguins` to see what happens on the full dataset!

```r
# Tạo ra dataset phụ có tên penguins_subset từ dataset chính đã khởi ban tạo ban đầu (penguins)
# creating a random subset of the penguins dataset
set.seed(42)

penguins_subset <- penguins %>%
  sample_n(50)  
  # another dplyr function!
  # một hàm khác của thư viện dplyr
  # Với hàm sample_n(x), ở dây ta chọn ngẫu nhiên từ toàn bộ dataset lấy 50 mẫu quan sát

# Xem lại dataset phụ này
penguins_subset
```

```r
# Chạy lại hàm arrange() trên dữ liệu ngẫu nhiên lấy từ 50 lần quan sát (dạng kí tự ~character)
# let's re-run the arrange() function on character data in the penguins_subset data

penguins_subset %>%
  arrange(Species)
```

### Ta có thể lồng hàm giảm dần là desc() bên trong hàm arrange()
Điều này có nghĩa hàm arange() mặc định là sắp xếp theo chiều tăng dần như chúng ta quan sát thấy ở trên.


```r
# Xắp xếp lại dữ liệu dạng số theo chiều giảm dần 
# numeric data arranged in descending order

penguins_subset %>%
  arrange(desc('Culmen Length (mm)'))
  # Lưu ý: như đã nói ở trên biến hay tên cột "Culmen Length (mm)" 
  #có khoảng trống nên khi khai báo cần phải để trong " " hay ' '
```

```r
# Xắp xếp lại dữ liệu dạng kí tự theo chiều giảm dần - ngược chiều bảng kí tự alpha beta
# character data arranged in descending - reverse alphabetical - order

penguins_subset %>%
  arrange(desc(Species))
```

## Thêm chức năng với hàm filter()

`filter()` có thể là 1 trong những hàm thường được sử dụng nhất bởi nóchota  nhìn  kĩ hơn vào dữ liệu một cách nhanh chóng và dễ dàng. Tính linh hoạt của hàm  `filter()` cho phép ta thêm 1 hay nhiều câu lệnh logic với các điều kiện khác nhau.

```r
# Lọc dữ liệu với 1 điều kiện dữ liệu dạng số
# filter with a single numeric condition

penguins_subset %>%
  filter('Culmen Depth (mm)' > 16.2)
```

```r
# Lọc dữ liệu với 1 điều kiện dữ liệu dạng kí tự (character)

# filter with a single character condition

penguins_subset %>%
  filter(Island == "Dream")
```

## Với câu lệnh/hàm mutate()
Chúng ta có thể dễ dạng tạo ra một cột mới (biến mới) bằng câu lệnh `mutate()`, nhưng khi ta làm việc mở rộng với notebook này, có thể sử dụng thêm hàm `group_by()` và sau đó dùng hàm `mutate()`!

```r
# Trong ví này, ta sẽ biến đổi đơn vịđo  lường từ grams sang dạng pound 
# Ta sẽ đặt tên biến mới là body_weight_pounds
# Và nó sẽ được thêm vào phía ngoài cùng bên  phải của dataframe hiện có 
# converting grams to pounds
# notice how the order of our columns stays the same, and the new column, body_weight_pounds, gets placed at the 
# far right of the dataframe. what function could we use to change this order?
penguins_subset %>% 
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
    # như đã nói nhiều lần là tên biến có vấn đề  nên ta cần đổi tên trước khi biến đổi dữ liệu
  mutate(body_weight_pounds = Body_Mass_g / 453.59237) %>%
  head()
```

```r
# Nếu chúng ta muốn kết hợp 4 hàm là rename(), select(), mutate(), và head()
# Ta sẽ làm như sau. Các bạn có thể tham khảo thêm tài liệu ở link bên dưới notebook này 
# OK I wanted to show you how to combine select and mutate
# what do you think the everything() function might do? confirm your guess by looking at the documentation (linked at 
# the end of the notebook).

penguins_subset %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  mutate(body_weight_pounds = Body_Mass_g / 453.59237) %>%
  select(Species, Body_Mass_g, body_weight_pounds, everything()) %>%
  head()
```

## Tổng hợp kết quả bằng câu lệnh  summarise(), kết hợp với hàm group_by()

Chúng ta có thể sử dụng hàm `summarise()` hay `summarize()` để tổng hợp kết quả lại từ dữ liệu. 
Khi kết hợp với hàm `group_by()` chúng ta có thể có kết quả chi tiết hơn từ dữ liệu toàn bộ dữ liệu.

```r
# Tổng kết trong lượng trung bình của chim cánh cụt theo đơn vị ddo lường là grams
# summarising the average body mass of penguins, in grams

penguins_subset %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  summarise(avg_body_mass = mean(Body_Mass_g))
```

```r
# Giờ hãy thử tính giá trị trung bình trọng lượng cơ thể theo grams của chim cánh cụt trong nghiên cứu
# since we're now summarising our data we can go ahead and use the full dataframe, since the printout will be reasonably-sized

penguins %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  summarise(avg_body_mass = mean(Body_Mass_g))
```

## HAAAAAAAA NAs!
`NA` viết tắt của "Not Available," nghĩa là missing data hay dữ liệu bị khuyết thiếu.
Tại sao kết quả in ra lại là NA? Và làm thế nào để xử lí vấn đề này?
Có rất nhiều cách xử lí vấn đề khuyết thiếu của dữ liệu trong R:
1. Có thể dùng hàm `na.rm = TRUE` ==> na = NA, rm = remove (loại bỏ), TRUE nghĩa là đúng. Tóm lại, câu lệnh nghĩa là hãy loại bỏ các quan sát khuyết thiếu
2. Có thể dùng hàm `filter()` từ thư viện `dplyr` 
3. Hay dùng hàm  `drop_na()` từ thư viện `tidyr` 


```r
# Tính trọng lượng trung bình cơ thể của chim, sau khi loại bỏ các dữ liệu khuyết thiếu 
# bằng hàm `na.rm = TRUE`
# summarising body mass on the entire penguins dataset while removing NA values from the calculation

penguins %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  summarise(avg_body_mass = mean(Body_Mass_g, na.rm = TRUE))
```

```r
# Giờ ta sẽ nhóm kết quả trọng lượng trung bình tính cơ thể theo grams theo loài
# now let's use the grouping function, group_by(), to look at the average body mass of penguins, in grams,
# by species

penguins %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  group_by(Species) %>%
  summarise(avg_species_body_mass = mean(Body_Mass_g, na.rm = TRUE)) 
```

```r
# Hãy tính trọng lượng cơ thể trung bình theo các đảo
# now let's calculate the average body mass by species AND island

penguins %>%
  rename(Body_Mass_g = 'Body Mass (g)') %>% 
  group_by(Species, Island) %>%
  summarise(avg_species_body_mass = mean(Body_Mass_g, na.rm = TRUE)) 
```
